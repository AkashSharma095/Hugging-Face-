```python
from transformers import AutoTokenizer, AutoModel, pipeline

tokenizer = AutoTokenizer.from_pretrained("avichr/heBERT_sentiment_analysis")
model = AutoModel.from_pretrained("avichr/heBERT_sentiment_analysis")
```

```python
import torch
first = torch.randn(932)
first = first.view(first.size(0),-1)
```

```python
import torch
from transformers import pipeline

# Assuming 'first' is a PyTorch tensor or string
# Convert tensors to lists of integers
if isinstance(first, torch.Tensor):
    first_list = first.view(first.size(0), -1).tolist()
    first_sentences = [str(item) for item in first_list]
elif isinstance(first, str):
    first_sentences = [first]
else:
    raise ValueError("Unsupported type for 'first'. Must be a PyTorch tensor or string.")

# Join the list of strings into a single paragraph
first_paragraph = ' '.join(first_sentences)

# Create the sentiment analysis pipeline
sentiment_analysis = pipeline(
    "sentiment-analysis",
    model="avichr/heBERT_sentiment_analysis",
    return_all_scores=True
)

# Split the paragraph into chunks of a certain length (e.g., 512)
max_chunk_length = 512
chunks = [first_paragraph[i:i + max_chunk_length] for i in range(0, len(first_paragraph), max_chunk_length)]

# Perform sentiment analysis on each chunk
results = [sentiment_analysis(chunk) for chunk in chunks]

print(results)

```

```python
from transformers import pipeline
from tqdm import tqdm

# Maximum sequence length supported by the model
max_seq_length = 512  

# Initialize the sentiment analysis pipeline
sent_pipeline = pipeline("sentiment-analysis")

# Initialize an empty dictionary to store sentiment analysis results
sentiment_results = {}

# Define the texts
texts = {'first': first, 'second': second}

# Perform sentiment analysis for each text
for text_id, text in texts.items():
    # Truncate the text if it exceeds the maximum sequence length
    if len(text) > max_seq_length:
        text = text[:max_seq_length]

    # Perform sentiment analysis
    sentiment_result = sent_pipeline(text)
    sentiment_results[text_id] = sentiment_result

# Display the sentiment analysis results
for text_id, result in sentiment_results.items():
    print(f"Text ID: {text_id}, Sentiment: {result}")
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] 
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

sentences = first
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-mpnet-base-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-mpnet-base-v2')
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')
with torch.no_grad():
    model_output = model(**encoded_input)
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)

print("Sentence embeddings:")
print(sentence_embeddings)
```
